```python
#create a separate list with even and odd number of size 10 and perform element wise addidion operation

even = [i for i in range(1,21) if i%2 == 0]
odd = [i for i in range(1,21) if i%2 !=0 ]
add = [i+j for i,j in zip(even,odd)]
add
```

```python
even = [i for i in range(1,21) if i%2 == 0]
odd = [i for i in range(1,21) if i%2 !=0 ]
add = []
for i in range(10):
    add.append(even[i] + odd[i])
print(add)
```

```python
""""1. write a list comprehension operation for genarating all the list of alphabetical characters in upper case and lower case.
2. write a code for pallindrome by using while statement
3. write a code to sort your string in alphabetical order. //don't use sort function """
```

```python
import string
s = [i for i in string.ascii_lowercase]
print(s, end=" ")

a = [i for i in string.ascii_uppercase]
print(a, end = " ")
```

```python
s = input("Enter a word to check for pallindrome: ")
ns = ""
n = len(s) - 1
while n != -1:
    ns = ns + s[n]
    n+=1
if ns==s:
    print("pallindrome")
else:
    print("not a palindrome")
```
